## Matriz de correlación

```python
variables_numericas = titanic.select_dtypes(include='number')
matriz_correlacion = variables_numericas.corr().round(2)
matriz_correlacion
```

## MAPA DE CALOR

```python
plt.imshow(matriz_correlacion, cmap="coolwarm", interpolation="nearest")
plt.colorbar(label="Coeficiente de correlación")

plt.xticks(ticks=np.arange(len(matriz_correlacion.columns)),
           labels=matriz_correlacion.columns,
           rotation=45, ha="right")
plt.yticks(ticks=np.arange(len(matriz_correlacion.columns)),
           labels=matriz_correlacion.columns)

for i in range(len(matriz_correlacion.columns)):
    for j in range(len(matriz_correlacion.columns)):
        valor = matriz_correlacion.iloc[i, j]
        plt.text(j, i, f"{valor:.2f}", ha="center", va="center", color="black")
        
plt.title('Correlación entre características (Features)')
plt.tight_layout()
plt.show()
```

```python
plt.imshow(matriz_correlacion, cmap="YlGnBu", interpolation="nearest")

plt.colorbar(label="Coeficiente de correlación")

plt.xticks(ticks=np.arange(len(matriz_correlacion.columns)),
           labels=matriz_correlacion.columns,
           rotation=45, ha="right")
plt.yticks(ticks=np.arange(len(matriz_correlacion.columns)),
           labels=matriz_correlacion.columns)

for i in range(len(matriz_correlacion.columns) + 1):
    plt.axhline(i - 0.5, color='white', lw=1)
    plt.axvline(i - 0.5, color='white', lw=1)

for i in range(len(matriz_correlacion.columns)):
    for j in range(len(matriz_correlacion.columns)):
        valor = matriz_correlacion.iloc[i, j]
        plt.text(j, i, f"{valor:.2f}", ha="center", va="center", color="black")

plt.title('Correlación entre características (Features)')
plt.tight_layout()
plt.show()
```

## APRENDIZAJE AUTOMÁTICO

```python
titanic_prep = titanic.drop(['Name','Ticket', 'Cabin','Embarked', 'PassengerId', 'FamilySize','Adulto'], axis=1)
titanic_prep.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
titanic_prep['Sex'] = label.fit_transform(titanic_prep['Sex'])
titanic_prep['CategoriaEdad'] = label.fit_transform(titanic_prep['CategoriaEdad'])
titanic_prep.head()
```

```python
titanic_prep.info()
```

```python
datos1 = titanic_prep.loc[:, ['Age', 'Fare']]
# KMEANS 1
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(datos1)
labels = kmeans.predict(datos1)
# visualization
plt.scatter(titanic_prep['Age'], titanic_prep['Fare'], c = labels)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
iner_list = np.empty(10)
for i in range(1,10):
 kmeans2 = KMeans(n_clusters=i)
 kmeans2.fit(datos1)
 iner_list[i] = kmeans2.inertia_
# iner_list = iner_list%100
# show the best number in graph
plt.plot(range(0,10), iner_list,'-')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
datos1 = titanic_prep.loc[:, ['Age', 'Fare']]
# KMEANS 1
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(datos1)
labels = kmeans.predict(datos1)
# visualization
plt.scatter(titanic_prep['Age'], titanic_prep['Fare'], c = labels)
```

```python
labels = kmeans.labels_
# Ver el número total de etiquetas diferentes (Equivale a Número de clusters encontrados)
num_clusters = np.unique(labels)
print("Etiquetas de clusters:",num_clusters)
print("Total de Registros:",len(kmeans.labels_))
print("Centroides:", kmeans.cluster_centers_)
print("Total de variables que forman parte de un centroide:",len(kmeans.cluster_centers_[0]))
```
