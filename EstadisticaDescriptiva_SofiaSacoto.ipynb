```python
mean_insulin = diabetes['Insulin'].mean()
median_insulin = diabetes['Insulin'].median()
mode_insulin = diabetes['Insulin'].mode()
print("Mean_insulin:",mean_insulin)
print("Median_insulin:",median_insulin)
print("Mode_insulin:",mode_insulin)
```

```python
mean_BMI = diabetes['BMI'].mean()
median_BMI = diabetes['BMI'].median()
mode_BMI = diabetes['BMI'].mode()
print("Mean_BMI:",mean_BMI)
print("Median_BMI:",median_BMI)
print("Mode_BMI:",mode_BMI)
```

He decidido escoger 2 variables:  
Insulin, variable cuantitativa tipo integer  
BMI, variable cuantitativa tipo float

La variable Insulin representa la concentración de insulina en sangre, el rango va desde el valor mínimo de 0 hasta el valor máximo de 846. Un rango así de amplio refleja alta diferencia entre los individuos. Por otro lado, la variable BMI, que quiere decir indice de masa corporal, mide la relación entre el peso y la altura de una persona con un rango desde el minimo 0 hasta el máximo de 67.1. Podemos ver que este rango nos sugiere que la mayoría de las personas están dentro del nivel de obesidad, ya que los valores normales de BMI se encuentran entre 18.5 y 24.9. 

Insulin presenta una media de 79.8 mucho mayor que su mediana de 30.5, esto sumado a la alta desviación estándar de 115.244 indica una distribución asimétrica y muy dispersa. Mientras tanto el BMI tiene media y mediana de 32 y una desviación estándar de 7.88, lo que muestra dentro de lo que cabe una distribución simétrica.
