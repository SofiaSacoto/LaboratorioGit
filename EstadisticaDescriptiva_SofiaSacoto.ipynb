## HISTOGRAMA

```python
import matplotlib.pyplot as plt
```

```python
# Histograma de insulina
plt.figure(figsize=(7,4))
plt.hist(diabetes['Insulin'], bins=20, color='blue', edgecolor='black')
plt.title("Distribución de Insulina")
plt.xlabel("Insulin")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Histograma de BMI
plt.figure(figsize=(7,4))
plt.hist(diabetes['BMI'], bins=20, color='green', edgecolor='black')
plt.title("Distribución de Índice de Masa Corporal (BMI)")
plt.xlabel("BMI")
plt.ylabel("Frecuencia")
plt.show()
```

Se aprecia una alta concentración de valores bajos en Insulin, indicando una fuerte asimetría positiva . En cambio, BMI se distribuye más simétricamente alrededor de 30–35.

## BOXPLOT

```python
# BOXPLOT de insulina
plt.figure(figsize=(6,4))
plt.boxplot(diabetes['Insulin'].dropna(), patch_artist=True, boxprops=dict(facecolor='blue'))
plt.title("Boxplot de Insulin")
plt.ylabel("Concentración")
plt.show()
```

```python
## BOXPLOT de BMI
plt.figure(figsize=(6,4))
plt.boxplot(diabetes['BMI'].dropna(), patch_artist=True, boxprops=dict(facecolor='green'))
plt.title("Boxplot de BMI")
plt.ylabel("Índice de Masa Corporal")
plt.show()
```

El boxplot de Insulin muestra múltiples valores atípicos y una mediana baja respecto al rango total, que representa una mayor dispersión. El boxplot de BMI tiene una forma más simétrica y achicada, indicando menor variabilidad.

## Matriz de correlación y mapa de calor

```python
matriz_corr = diabetes[['Insulin','BMI','Outcome']].corr().round(2)

plt.imshow(matriz_corr, cmap="YlGnBu", interpolation="nearest")
plt.colorbar(label="Coeficiente de correlación")
plt.xticks(np.arange(3), matriz_corr.columns)
plt.yticks(np.arange(3), matriz_corr.columns)

for i in range(3):
    for j in range(3):
        valor = matriz_corr.iloc[i, j]
        plt.text(j, i, f"{valor:.2f}", ha="center", va="center", color="black")

plt.title("Mapa de Calor - Correlación entre Insulin y BMI")
plt.tight_layout()
plt.show()
```

En conjunto, las correlaciones son bajas y positivas, por lo que no hay una relación lineal fuerte entre las variables, aunque el BMI parece tener una influencia un poco mayor sobre el resultado de diabetes que la insulina.
