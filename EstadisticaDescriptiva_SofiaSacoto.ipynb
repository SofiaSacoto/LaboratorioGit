## HISTOGRAMA

```python
import matplotlib.pyplot as plt
```

```python
resultado = diabetes['Outcome'].value_counts()
resultado.plot(kind='bar')
plt.title('Distribución de outcome')
plt.show()
```

```python
# Histograma de insulina
plt.figure(figsize=(7,4))
plt.hist(diabetes['Insulin'], bins=20, color='blue', edgecolor='black')
plt.title("Distribución de Insulina")
plt.xlabel("Insulin")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Histograma de BMI
plt.figure(figsize=(7,4))
plt.hist(diabetes['BMI'], bins=20, color='green', edgecolor='black')
plt.title("Distribución de Índice de Masa Corporal (BMI)")
plt.xlabel("BMI")
plt.ylabel("Frecuencia")
plt.show()
```

Se aprecia una alta concentración de valores bajos en Insulin, indicando una fuerte asimetría positiva . En cambio, BMI se distribuye más simétricamente alrededor de 30–35.

## BOXPLOT

```python
# BOXPLOT de insulina
plt.figure(figsize=(6,4))
plt.boxplot(diabetes['Insulin'].dropna(), patch_artist=True, boxprops=dict(facecolor='blue'))
plt.title("Boxplot de Insulin")
plt.ylabel("Concentración")
plt.show()
```

```python
## BOXPLOT de BMI
plt.figure(figsize=(6,4))
plt.boxplot(diabetes['BMI'].dropna(), patch_artist=True, boxprops=dict(facecolor='green'))
plt.title("Boxplot de BMI")
plt.ylabel("Índice de Masa Corporal")
plt.show()
```

El boxplot de Insulin muestra múltiples valores atípicos y una mediana baja respecto al rango total, que representa una mayor dispersión. El boxplot de BMI tiene una forma más simétrica y achicada, indicando menor variabilidad.

## Comparación con outcome

```python

b0 = diabetes[diabetes["Outcome"]==0]["BMI"]
b1 = diabetes[diabetes["Outcome"]==1]["BMI"]

plt.figure(figsize=(6,4))
plt.boxplot([b0, b1], labels=["Outcome=0","Outcome=1"], patch_artist=True)
plt.title("BMI por Outcome")
plt.ylabel("BMI")
plt.show()
```

El diagrama muestra que las personas con diagnóstico positivo de diabetes tienden a tener valores de BMI más altos que aquellas con diagnostico negativo. Además podemos observar que la mediana del grupo diabético se encuentra por encima de la mediana del grupo no diabético, lo puede sugerir que si existe una relación entre el BMI y la posibilidad de tener diabetes.

```python
i0 = diabetes[diabetes["Outcome"]==0]["Insulin"]
i1 = diabetes[diabetes["Outcome"]==1]["Insulin"]

plt.figure(figsize=(6,4))
plt.boxplot([i0, i1], labels=["Outcome=0","Outcome=1"], patch_artist=True)
plt.title("Insulin por Outcome")
plt.ylabel("Insulin (µU/mL)")
plt.show()
```

La mediana de insulina en el grupo con diagnóstico de diabetes positiva es ligeramente mayor que lo negativo, lo que sugiere que las personas con diabetes tienden a tener niveles de insulina más altos, aunque la diferencia no es muy marcada. También se debe recalcar que existe una gran variabilidad tal vez deibido a la presencia de valores extremos sugieren diferencias marcadas.
